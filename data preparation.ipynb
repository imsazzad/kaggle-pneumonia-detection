```python
train_annotation = pd.read_csv('stage_1_train_labels.csv')
```

```python
for index, row in train_annotation.iterrows():
    if(index > -1):
        patient_id = row['patientId']
        x_cordinate = row['x']
        y_cordinate = row['y']
        width = row['width']
        height = row['height']
        class_number = row['Target']
        
        #print(patient_id)
        import cv2
        full_img_path =  'stage_1_train_jpg_images/' + patient_id + ".jpg"
        patient_img = cv2.imread(full_img_path, cv2.IMREAD_UNCHANGED)
        #print(patient_img.shape)
        img_height = patient_img.shape[0]
        img_width = patient_img.shape[1]
        #img_channels = patient_img.shape[2]
        
        f = open("yolo_img_annotation/"+ patient_id + ".txt", "w")
        if(int(class_number) != 0):
            obj_class = int(class_number)
            x = (x_cordinate + width/2 )/img_width
            y = (y_cordinate + height/2) /img_height
            w = width/img_width
            h = height/img_height
            
            txt_to_write = obj_class -1, x, y, w, h
            print(txt_to_write)
            
            
            f.write(str(obj_class -1) + " " + str(x) +" " + str(y) + " " + str(w) +" "+ str(h))
            
            
        f.close() 
    #print()
```

```python
def prepare_train_test_txt_from_input_image_folder(input_image_folder , test_percentage = 20,
                                                   train_file = 'train.txt', test_file ='test.txt' ):
    import glob, os
    # Current directory
    #current_dir = os.path.dirname(os.path.abspath(__file__))
    #print(current_dir)
    
    current_dir = input_image_folder 
    #current_dir = 'images'
    # Percentage of images to be used for the test set
#     percentage_test = 20;
    # Create and/or truncate train.txt and test.txt
    file_train = open(train_file, 'w')  
    file_test = open(test_file, 'w')
    # Populate train.txt and test.txt
    counter = 1  
    index_test = round(100 / test_percentage)  
    for pathAndFilename in glob.iglob(os.path.join(current_dir, "*.jpg")):  
        title, ext = os.path.splitext(os.path.basename(pathAndFilename))
        if counter == index_test:
            counter = 1
            #print(counter)
            file_test.write(current_dir + "/" + title + '.jpg' + "\n")
        else:
            #print(counter)
            file_train.write(current_dir + "/" + title + '.jpg' + "\n")
            counter = counter + 1
```
